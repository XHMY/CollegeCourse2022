# 20190105 曾一凡 Selenium 爬虫作业

具体要求：

使用Selenium爬取51_job_python职位内容

1. 提取职位名称、公司名称、福利信息、学历要求
2. 将爬取结果存储到csv中

```python
from selenium import webdriver
import pandas as pd
```

## 一、启用驱动

```python
driver = webdriver.Chrome()
url = "https://www.51job.com/"
driver.get(url)
```

## 二、根据关键词在相关元素进行操作

```python
driver.find_element_by_id("kwdselectid").send_keys("Python")
driver.find_element_by_xpath("/html/body/div[3]/div/div[1]/div/button").click()
```

## 三、爬取职业名称

```python
jobs_name = [i.get_attribute('title') for i in driver.find_elements_by_xpath("//p//span[@title]")]
len(jobs_name) #查看长度确保没有漏爬数据
```

## 四、爬取公司名称

```python
companies_name = [i.get_attribute('text') for i in driver.find_elements_by_xpath("//div[@class='er']//a")]
len(companies_name) #查看长度确保没有漏爬数据
```

## 五、爬取福利信息

```python
message_block = []
message_block = [i for i in driver.find_elements_by_xpath("/html/body/div[2]/div[3]/div/div[2]/div[4]/div[1]/div/a")]
len(message_block) #查看长度确保没有漏爬数据
```

```python
welfare_block = [i.find_elements_by_xpath('p[3]') for i in driver.find_elements_by_xpath("/html/body/div[2]/div[3]/div/div[2]/div[4]/div[1]/div/a")]
len(welfare_block) #查看长度确保没有漏爬数据
```

```python
welfare = []
for i in welfare_block:
    if len(i)==0:
        welfare.append('')
    else:
        for j in i:
            welfare.append(j.get_attribute('title'))
len(welfare) #查看长度确保没有漏爬数据
```

## 六、爬取学历

```python
qualifications = [i for i in driver.find_elements_by_xpath('//div//a//p//span[@class="d at"]')]
len(qualifications)
```

```python
qualification = []
for i in qualifications:
    qualification.append(i.text)
```

```python
education = []
for edu in qualification:
    education.append(edu[-2:])
```

## 七、导入csv

```python
jobs_info = {'Jobs_name':jobs_name,'Companies_name':companies_name,'Welfare':welfare,'Education':education}
df = pd.DataFrame(jobs_info)
df
```

```python
df.to_csv('51job_python_selenium.csv',encoding='utf_8_sig') # 导出时要加上encoding='utf_8_sig'，否则直接打开此csv会直接乱码
```
