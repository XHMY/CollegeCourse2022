```python
import pandas as pd
```

```python
df = pd.read_excel("./NationGDP.xlsx")
df.head()
```

```python
df.sort_values("GDP", ascending=False, inplace=True)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #该语句解决图像中的“-”负号的乱码问题
color=['#66FFFF', '#66FFCC', '#66FF99', '#66FF66', '#99FF66', '#CCFF66', '#FFFF66', '#FFCC66', '#FF9966', '#FF6666']
```

```python
plt.barh(df.Country, df.GDP, color=color)
plt.show()
```

```python
plt.bar(df.Country, df.GDP,color=color)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df.GDP,
        labels=df.Country, # 设置饼图标签
        explode=(0, 0.2, 0, 0,0,0,0,0,0,0),
        autopct='%.2f%%',
        textprops={'fontsize':12,'color':'#000000'}
       )
plt.show()
```
