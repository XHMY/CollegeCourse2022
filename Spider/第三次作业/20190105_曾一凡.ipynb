```python
import requests
import re
import pandas as pd
```

```python
url = 'https://search.51job.com/list/000000,000000,0000,00,9,99,python,2,1.html?lang=c&postchannel=0000&workyear=99&cotype=99&degreefrom=99&jobterm=99&companysize=99&ord_field=0&dibiaoid=0&line=&welfare='
```

```python
r = requests.get(url)
print(r.text)
```

```python
pattern = r'{"type":.*?"job_name":"(?P<job>.*?)".*?"company_name":"(?P<company>.*?)".*?"providesalary_text":"(?P<salary>.*?)".*?"workarea_text":"(?P<area>.*?)".*?"companyind_text":"(?P<ind>.*?)".*?}'
result = re.finditer(pattern, r.text)
```

```python
df = pd.DataFrame([{r.group(k) for k in ("job", "company", "salary", "area", "ind")} for r in result])
df.to_csv('51job_python.csv')
df.head()
```
